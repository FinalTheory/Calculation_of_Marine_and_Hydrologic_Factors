## 北太平洋SST数据的EOF分解及分析

本次作业继续基于Ipython notebook呈现。由于不要求编写程序，因此使用eofs工具包来实现EOF分解；绘图部分则使用Python的Basemap工具箱。

### 数据处理

- 首先导入本次作业需要用到的工具：

```python
import numpy as np
from pylab import *
from netCDF4 import Dataset
from eofs.standard import Eof
import matplotlib.pyplot as plt
from IPython.display import Image
from mpl_toolkits.basemap import Basemap
```

- 然后读取数据，取出对应的经纬度。

```python
filename = 'HadISST_sst.nc'
ncin = Dataset(filename, 'r')
lons = ncin.variables['longitude'][:]
lats = ncin.variables['latitude'][:]
```

- 简单查看一下这些数据的属性，可以看出维度排列顺序是“时间-纬度-经度”。

```python
print ncin.variables['sst'].shape
print lons.shape
print lats.shape
```

- 下面开始提取目标范围内的经纬度，并合并为一个矩阵，再踢除掉异常值。
- 因为太平洋这部分的数据需要从矩阵左右两头拼接，所以使用了这种比较麻烦的做法。

```python
sst1 = ncin.variables['sst'][1200:1440, 69:8:-1, 309:360]
sst2 = ncin.variables['sst'][1200:1440, 69:8:-1, 0:90]
sst = concatenate((sst1, sst2), axis=2)
lat = lats[69:8:-1]
lon = hstack((lons[309:360], lons[0:90] + 360))
sst[sst < -100] = NaN
print sst1.shape
print sst2.shape
print sst.shape
```

- 查看一下温度的取值范围：

```python
print sst[~isnan(sst)].max()
print sst[~isnan(sst)].min()
```

- 开始求距平值以用于EOF分解，并查看一下最大值以及最小值。

```python
ssta = np.zeros(sst.shape)
aver = np.zeros((sst.shape[1], sst.shape[2]))
for t in range(0, 12):
    mask = np.arange(0, sst.shape[0]) % 12 == t
    aver[:, :] = squeeze(sum(sst[mask, :, :], axis=0)) / find(mask).shape[0];
    for i in find(mask):
        ssta[i, :, :] = sst[i, :, :] - aver[:, :]
        
print ssta[~isnan(ssta)].max()
print ssta[~isnan(ssta)].min()
```

- 然后调用basemap作图，查看SSTA的具体分布情况：

```python
fig = plt.figure(figsize=(12,7))
m = Basemap(projection='mill',llcrnrlat=20,urcrnrlat=80,\
            llcrnrlon=130,urcrnrlon=270,resolution='l')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='white')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,91.,20.))
m.drawmeridians(np.arange(-180.,181.,20.))
m.drawmapboundary(fill_color='white')
m.drawcountries(linewidth=1.0)
x, y = m(*np.meshgrid(lon, lat))
cs = m.contourf(x, y, ssta[0,:,:], 15, cmap=plt.cm.jet)
cbar = m.colorbar(cs, location='bottom')
cbar.set_label('SSTA $^{\circ}$C', fontsize=15)
plt.title('SSTA of North Pacific', fontsize=20)
plt.show()
```

### EOF分解
- 下面正式开始进行EOF分析，这里直接求前10个模态及其时间系数。

```python
solver = Eof(ssta)
eofs = solver.eofs(neofs=10)
pcs = solver.pcs(npcs=10, pcscaling=0)
```

```python
weights = solver.eigenvalues(10) / sum(solver.eigenvalues())
summ = cumsum(weights)
print '各个模态方差贡献率以及对应的累计方差贡献率：'
for idx in range(len(weights)):
    print 'EOF %d:\t%.4f\t\t%.4f' % (idx + 1, weights[idx], summ[idx])
```

- 可以看出前5个模态是最为主要的部分，其总贡献超过50%，于是直接将前5个模态以及其对应的时间系数画出来：

```python
for i in range(5):
    fig, axis = plt.subplots(1,2, figsize=(20, 6))
    m = Basemap(projection='mill',llcrnrlat=20,urcrnrlat=80,llcrnrlon=130,urcrnrlon=270,resolution='l', ax=axis[0])
    m.drawcoastlines()
    m.fillcontinents(color='coral',lake_color='white')
    # draw parallels and meridians.
    m.drawparallels(np.arange(-90.,91.,20.),labels=[1,0,0,0],fontsize=10)
    m.drawmeridians(np.arange(-180.,181.,20.),labels=[0,0,0,1],fontsize=10)
    m.drawmapboundary(fill_color='white')
    m.drawcountries(linewidth=1.0)
    x, y = m(*np.meshgrid(lon, lat))
    cs = m.contourf(x, y, eofs[i,:,:].squeeze(), 20, cmap=plt.cm.jet)
    cbar = m.colorbar(cs,location='bottom',pad="8%")
    axis[0].set_title('EOF ' + str(i + 1), fontsize=20)
    axis[1].plot(pcs[:, i])
    axis[1].set_xlabel('Months since 1971', fontsize=15)
    axis[1].set_title('PC ' + str(i + 1) + ' Time Series', fontsize=20)
    plt.show()
```

### 分析与讨论

#### 第一模态

- EOF第一模态为温度场总方差贡献了24.9%。从图中来看，太平洋中西部被一片强正值所占据，可以明显地看出有两个中心；而北美沿岸地区以及太平洋东部体现为负值区。考虑到SSTA场的实际变化取决于时间系数与空间场的乘积，因此观察时间系数的变化，可以看出北太平洋西、中部海温与东部及沿海地区海温是反相关系，并且负值区与北太平洋西风漂流区吻合。
- 观察PDO指数在给定的时间段，与第一模态的时间系数有着较好的对应关系，这是因为PDO指数就是这样定义的：The Pacific Decadal Oscillation (PDO) Index is defined as the leading principal component of North Pacific monthly sea surface temperature variability (poleward of 20N for the 1900-93 period).
- 由时间序列可知：1979年初至1989年初，时间序列指数基本为正值，说明图中蓝色区域海表面温度低于红色区域海表面温度，即北太平洋西、中部海温低于东部海温；1971年至1978年底，以及1990年前后，时间序列指数基本为负值，说明中蓝色区域海表面温度高于红色区域海表面温度，即北太平洋西、中部海温高于东部海温。此分析和历史资料相吻合。
- EOF第一主要模态可以反映北太SST和PDO有关的大尺度分布特征，因此这种模态分布是全球海洋与大气相互作用的一个重要组成部分，它是北太平洋海温非季节变化的重要的形态。

#### eofs工具介绍  

eofs is a Python package for EOF analysis of spatial-temporal data, some of its key features are:

- Suitable for large data sets: computationally efficient for the large output data sets of modern climate models.
- Transparent handling of missing values: missing values are removed automatically during computations and placed back into output fields.
- Automatic metadata: metadata from input fields is used to construct metadata for output fields (requires the cdms2 module from CDAT or the iris data analysis package).
- No Compiler required: a fast implementation written in pure Python using the power of numpy, no Fortran or C dependencies.

#### basemap工具介绍

- The matplotlib basemap toolkit is a library for plotting 2D data on maps in Python. It is similar in functionality to the matlab mapping toolbox, the IDL mapping facilities, GrADS, or the Generic Mapping Tools. PyNGL and CDAT are other libraries that provide similar capabilities in Python.

- Basemap does not do any plotting on it’s own, but provides the facilities to transform coordinates to one of 25 different map projections (using the PROJ.4 C library). Matplotlib is then used to plot contours, images, vectors, lines or points in the transformed coordinates. Shoreline, river and political boundary datasets (from Generic Mapping Tools) are provided, along with methods for plotting them. The GEOS library is used internally to clip the coastline and polticial boundary features to the desired map projection region.

- Basemap provides facilities for reading shapefiles.

- Basemap is geared toward the needs of earth scientists, particular oceanographers and meteorologists. I originally wrote Basemap to help in my research (climate and weather forecasting), since at the time CDAT was the only other tool in python for plotting data on map projections. Over the years, the capabilities of Basemap have evolved as scientists in other disciplines (such as biology, geology and geophysics) requested and contributed new features.
