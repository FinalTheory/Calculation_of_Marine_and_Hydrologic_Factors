## 异常数据处理
- 导入所有需要的包

```python
import os
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

- 定义一个函数来对文本进行简单的解析，略去无用的注释信息仅保留可用数据

```python
def parse_txt(file_name):
    if not (os.path.exists(file_name) and os.path.isfile(file_name)):
        print "File doesn't exist!"
        return
    FILE = open(file_name, 'r')
    lines = FILE.readlines()
    aver = 0
    for line in lines:
        aver += len(line)
    aver /= len(lines)
    new_file = open(file_name + '.raw_data', 'w')
    for line in lines:
        if len(line) > aver:
            new_file.write(line)
    FILE.close()
    new_file.close()
    return file_name + '.raw_data'
```

- 调用函数以处理原始数据文件

```python
filename = parse_txt('KM9708new.dat')
```

### 潮汐资料调和分析工具使用说明
1. 按照指定格式编辑配置文件constituent.ini，输入选定的分潮；已将课本P87页表格中所有分潮写入该文件。
1. 用**数据中间时刻的对应时间**，以及**需要用于预报的分潮名称**来调用数据读取工具**getData**，获取这些参数：主分潮数量(P)、随从分潮数量(Q)、角速度(sigma)、初相角(V)、交点因子(f)、振幅比(kappa)、迟角差(alpha)、主分潮和随从分潮的对应关系(corr)、选取用来做预报的分潮编号(mask)、分潮数字编号与名称的对应关系(record)。
    1. 在上述过程中，用于预报的分潮名称可以省略，此时默认使用四大分潮和两个浅水分潮进行预报，如下命令所示：
        > P, Q, sigma, V, f, kappa, alpha, corr, mask, record = get_variables(middle_date)
    1. 如果输入一个空列表，则会自动使用所有分潮进行潮位预报，如下命令所示：
        > P, Q, sigma, V, f, kappa, alpha, corr, mask, record = get_variables(middle_date, [])
    1. 如果输入其他有效参数，则使用指定的分潮进行潮位预报，如下命令所示：
        > P, Q, sigma, V, f, kappa, alpha, corr, mask, record = get_variables(middle_date, ['m_2', 's_2', 'k_1', 'o_1', 'm_4', 'ms_4'])
    1. 最后还有一个隐藏参数，用于指定所计算海区的$\Delta g$取值，如下命令所示：
        > P, Q, sigma, V, f, kappa, alpha, corr, mask, record = get_variables(middle_date, ['m_2', 's_2', 'k_1', 'o_1', 'm_4', 'ms_4'], delta_g = 50.)
1. 以这些参数调用调和分析计算模块，用以计算振幅$H$、迟角$g$和平均海面$S_{0}$，如下命令所示：
    > H, g, S_0 = Tides.analyze(P, zeta, sigma, V, f, kappa, alpha, corr)
1. 继续使用上述参数和调和常数调用潮位预测模块，用以回报潮位过程曲线。此时需要指定开始与结束的时间点（均为相对于原始数据中间时刻的偏移）和数据的时间间隔。
    1. 下述命令回报的范围为数据范围，即从相对于中间时刻第-371个数据到相对于中间时刻的第371个数据，时间间隔为1.0小时：
        > zeta_predict = Tides.predict(-371, 371, sigma, V, f, H, g, S_0, mask, 1.)
    2. 下面的命令由于时间间隔改为了1/60小时，所以数据个数就需要对应乘以60：
        > zeta_predict_minute = Tides.predict(-371 \* 60, 371 \* 60, sigma, V, f, H, g, S_0, mask, 1./60.)

### 使用方法示意

- 下面使用“KM9708.dat”中给出的真实测量数据进行调和分析和潮位回报。
- 首先去掉最后一个观测值，将观测数据个数取为奇数。

```python
zeta = data_real[0:-1]
print zeta.size
```

- 此时中间时刻对应的数据应该是第(743-1)/2个，即第371个。
- 于是利用datetime模块计算一下这个时间对应的是什么时候，可以看出对应的是8月16号的中午12点。
- 不过需要修正到格林尼治时间，做法是直接减去八个小时即可。
- 这里只是给出一个处理日期间隔的简单方案而已，当然更简便的做法是用Wolfram Alpha。

```python
import dateutil
date = dateutil.parser.parse('1997.8.1 1:00')
#修正到对应的格林尼治时间，即减去八个小时
middle_date = (date + datetime.timedelta(hours = 371) - datetime.timedelta(hours=8)).ctime()
print middle_date
```

- **进行调和分析并计算相关系数**

```python
#导入分潮数据读取模块和调和分析运算模块
import Tides
from getData import *
#输入中间时刻，计算天文要素
P, Q, sigma, V, f, kappa, alpha, corr, mask, record = get_variables(middle_date)
#进行调和分析，计算调和常数
H, g, S_0 = Tides.analyze(P, zeta, sigma, V, f, kappa, alpha, corr)
#带入调和常数进行回报
zeta_predict = Tides.predict(-371, 371, sigma, V, f, H, g, S_0, mask, 1.)
#回报逐分钟的潮位用以计算高低潮时和潮高
zeta_predict_minute = Tides.predict(-371*60, 371*60, sigma, V, f, H, g, S_0, mask, 1./60.)
#同图绘制回报过程曲线与实测过程曲线
show_diff(zeta, zeta_predict);
#计算回报的过程曲线与真实测量值的相关系数，其数值相当接近1，说明回报结果正确
print '相关系数：%.5f' % xcorr(zeta, zeta_predict, maxlags=1)[1][1]
print '平均海面：%.2fmm' % S_0
print '分潮及其对应的H、g：'
for k in record.keys():
    print '分潮名称: %s\t\t振幅: %.2fmm\t\t迟角: %.2frad' % (record[k], H[k-1], g[k-1])
```

### 计算高低潮的潮时、潮高、平均潮差、余水位

由于这部分的计算原理实在是比较简单，因此直接用Python实现：

```python
def print_max_min(data, time):
    ssum1 = 0.
    ssum2 = 0.
    num1 = 0
    num2 = 0
    t = dateutil.parser.parse(time)
    for i in range(1, data.size - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            print '高潮时: %s\t潮位:%dmm' % ( (t + timedelta(minutes=i)).strftime('%c'), data[i] )
            ssum1 += data[i]
            num1 += 1
        elif data[i] < data[i - 1] and data[i] < data[i + 1]:
            print '低潮时: %s\t潮位:%dmm' % ( (t + timedelta(minutes=i)).strftime('%c'), data[i] )
            ssum2 += data[i]
            num2 += 1
    print '平均潮差：%.2fmm' % ( ssum1 / float(num1) - ssum2 / float(num2)  )

print_max_min(zeta_predict_minute, '1997.8.1 1:00')
print '绘制余水位：'
fig, axes = plt.subplots(figsize=(18,4));
axes.plot(zeta - zeta_predict);
```

### 结论与分析
**通过计算潮型数，可知A=0.274，KM站为半日潮港。从上述高低潮时中也能明显得出此结论**。  
此外，回报过程曲线与真实测量值的对比中可以看出，从约第400个数据开始的一段时间里，真实值相对于回报值明显高出一段距离，即存在异常的增水。  
这段时间大概对应于97年8月17号到18号左右，查阅资料后发现这段时间浙江地区有台风登录，所以可以解释水位的这一异常变化。

## 工具介绍

### Python

Python作为一种高度成熟且开源免费的科学计算语言，在海洋学领域的应用尚不广泛，这里对其进行简单的介绍。  
主要特点：

- 语法简明清晰：代码直观且可读性强，项目规模的增大不会显著降低其可维护性。
- 表现能力出色：实现功能所需的代码量很短，这可以有效减少BUG，易于编写。

技术细节：

- 动态类型：如同MATLAB一样定义变量时不需要声明其类型，同样也不需要声明函数参数、返回值的类型。
- 自动垃圾回收：如同Java一样不需要为对象和数组显式地分配或回收内存，不会导致任何内存泄漏的问题。
- 解释型语言：不需要对代码进行编译，直接解释执行即可（执行时会自动编译成字节码，类似MATLAB的JIT技术）。

优势：

- 将Python引入海洋学研究中的最大优势在于简化了编程的复杂度，避免将精力消耗在语言细节上。
- 具有卓越的语法设计，相比于MATLAB等简陋的过程化编程，其现代化的面向对象思想以及设计完善的包管理机制最大限度地提升了代码的可重用性和可维护性。
- 本身还具有胶水语言的特性，可以很方便地与C、Fortran等高性能计算语言进行混合编程，不仅弥补了自身运行效率的不足，同时也简化了编程工作——可以直接复用。
- 拥有数目巨大的各类库函数，以及开源社区的大力支持，可以用它实现几乎一切编程需求——数据读取、处理、绘图、编写用户图形界面等所有工作都可以用Python完成。

缺陷：

- 相比于C、Fortran等编译型语言，Python等动态解释型语言的最大缺陷就是运行效率较差。不过上面已经提到过，混合编程可以弥补这一缺陷。
